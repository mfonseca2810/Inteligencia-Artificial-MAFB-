## 1. Carga y Visualización Inicial
En esta sección cargamos el dataset, inspeccionamos dimensiones, tipos de datos y valores nulos.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
pd.set_option('display.max_columns', 100)

# Cargar datos (ajusta la ruta si es necesario)
df = pd.read_csv('train.csv')

print('Dimensiones del dataset:', df.shape)
display(df.head())
print('\nInformación general:')
df.info()
print('\nNulos (primeros 20):')
display(df.isnull().sum().head(20))
```

## 2. Diagnóstico de Valores Faltantes
Calculamos un reporte ordenado por columnas con valores nulos.

```python
missing_values = df.isnull().sum().sort_values(ascending=False)
missing_values = missing_values[missing_values > 0]
missing_percentage = (missing_values / len(df)) * 100
missing_report = pd.DataFrame({'Valores Nulos': missing_values,
                               '% del Total': missing_percentage.round(2)})
print('--- Reporte de valores nulos ---')
display(missing_report)
```

## 3. Limpieza de Datos
Decisiones principales:
- Eliminar columnas con >80% de nulos: `PoolQC`, `MiscFeature`, `Alley`, `Fence`.
- Imputar `MasVnrType` (moda), `MasVnrArea` (mediana), `Electrical` (moda).
- Imputar atributos de sótano con **'Sin Sotano'** en categóricas.
- Imputar atributos de garaje con **'Sin Garaje'** en categóricas; `GarageYrBlt` con 0.
- Imputar `FireplaceQu` con **'Sin Chimenea'**.
- Imputar `LotFrontage` con la **mediana por `Neighborhood`**.

```python
df_clean = df.copy()
cols_drop = ['PoolQC', 'MiscFeature', 'Alley', 'Fence']
df_clean.drop(columns=cols_drop, inplace=True)

# Imputaciones simples
df_clean['MasVnrType'].fillna(df_clean['MasVnrType'].mode()[0], inplace=True)
df_clean['MasVnrArea'].fillna(df_clean['MasVnrArea'].median(), inplace=True)
df_clean['Electrical'].fillna(df_clean['Electrical'].mode()[0], inplace=True)

# Sótano
basement_cols = ['BsmtQual', 'BsmtCond', 'BsmtExposure', 'BsmtFinType1', 'BsmtFinType2']
for col in basement_cols:
    df_clean[col].fillna('Sin Sotano', inplace=True)

# Garaje
garage_cat_cols = ['GarageType', 'GarageFinish', 'GarageQual', 'GarageCond']
for col in garage_cat_cols:
    df_clean[col].fillna('Sin Garaje', inplace=True)
df_clean['GarageYrBlt'].fillna(0, inplace=True)

# Chimenea
df_clean['FireplaceQu'].fillna('Sin Chimenea', inplace=True)

# LotFrontage por vecindario
df_clean['LotFrontage'] = df_clean.groupby('Neighborhood')['LotFrontage'].transform(lambda x: x.fillna(x.median()))

print('Nulos totales luego de la limpieza:', int(df_clean.isnull().sum().sum()))
```

## 4. Análisis Descriptivo
Obtenemos resúmenes estadísticos y visualizamos distribuciones clave.

```python
# Resumen estadístico
display(df_clean.describe())

# Distribución de la variable objetivo
plt.figure(figsize=(8,5))
sns.histplot(df_clean['SalePrice'], kde=True, bins=30)
plt.title('Distribución de SalePrice')
plt.xlabel('SalePrice')
plt.ylabel('Frecuencia')
plt.show()

# Categóricas importantes
plt.figure(figsize=(8,5))
sns.countplot(x='OverallQual', data=df_clean)
plt.title('Distribución de OverallQual (Calidad General)')
plt.show()

plt.figure(figsize=(12,5))
sns.countplot(x='Neighborhood', data=df_clean, order=df_clean['Neighborhood'].value_counts().index)
plt.title('Distribución de Neighborhood (Vecindarios)')
plt.xticks(rotation=90)
plt.show()

plt.figure(figsize=(6,5))
sns.countplot(x='GarageCars', data=df_clean)
plt.title('Distribución de GarageCars (Capacidad de Garaje)')
plt.show()
```

## 5. Análisis Exploratorio (EDA)
Calculamos correlaciones con `SalePrice`, graficamos un heatmap, scatter plots y box plots.

```python
# Matriz y ranking de correlación
corr_matrix = df_clean.corr(numeric_only=True)
corr_with_price = corr_matrix['SalePrice'].sort_values(ascending=False)
print('Top 10 variables numéricas más correlacionadas con SalePrice:')
display(corr_with_price.head(10))

plt.figure(figsize=(12,8))
sns.heatmap(corr_matrix, cmap='coolwarm', center=0)
plt.title('Matriz de correlación')
plt.show()

# Scatter plots de las 3 más correlacionadas (excluyendo SalePrice)
top_corr = corr_with_price.index[1:4]
for col in top_corr:
    plt.figure(figsize=(8,5))
    sns.scatterplot(x=df_clean[col], y=df_clean['SalePrice'])
    plt.title(f'Relación entre {col} y SalePrice')
    plt.xlabel(col)
    plt.ylabel('SalePrice')
    plt.show()

# Box plots con variables categóricas
plt.figure(figsize=(8,5))
sns.boxplot(x='OverallQual', y='SalePrice', data=df_clean)
plt.title('SalePrice vs OverallQual')
plt.show()

plt.figure(figsize=(8,5))
sns.boxplot(x='GarageCars', y='SalePrice', data=df_clean)
plt.title('SalePrice vs GarageCars')
plt.show()
```

## 6. Conclusión Final
A partir del análisis:

- **OverallQual** (calidad general) es el predictor más fuerte del precio.
- **GrLivArea** (área habitable) y **TotalBsmtSF** (área total de sótano) muestran fuerte relación positiva.
- **GarageCars** y **GarageArea** contribuyen de forma apreciable.
- Variables de contexto como **Neighborhood** también influyen, aunque con efecto menor.

**Resumen:** Las casas con mayor calidad de construcción, mayor área habitable y mejor capacidad de garaje tienden a venderse a precios más altos.
